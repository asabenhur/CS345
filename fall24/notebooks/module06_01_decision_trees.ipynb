# Decision Trees

Decision trees are a classic machine learning technique that has been very popular because of its intuitive nature and fast training algorithms.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Decision trees on the iris dataset

The iris dataset is an excellent dataset to introduce decision trees.  We will visualize it to remind ourselves of its characteristics.

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y  = iris.target
iris.feature_names
```

```python
def scatter_with_legend(X, y, ax, class_names) :
    colors = ['ob', 'or', 'oc']
    for c in range(len(class_names)):
        idx = (y==c)
        ax.plot(X[idx,0], X[idx,1], colors[c], 
                label = class_names[c], alpha=0.5)
    ax.legend()
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(6,3))
scatter_with_legend(X[:,[0, 1]], y, ax1, iris.target_names)
scatter_with_legend(X[:,[2, 3]], y, ax2, iris.target_names)
ax1.set_title('sepal length/width')
ax2.set_title('petal length/width');
```

From these plots we can see that there are simple rules based on the values of single features that can provide good accuracy.  For example, if petal length is less than 2.5, we would classify an iris as belonging to the species iris setosa.

Decision trees capture the very intuitive way in which we as humans might make decisions, as a set of rules.  As an example consider the following very simple decision tree that provides high classification accuracy between Iris versicolor and Iris virginica:

### The decision tree training algorithm

The decision tree training algorithm iteratively splits the data into two subsets based on the value of a single feature.
At each step the algorithm picks the feature that leads to the greatest "purity" of the resulting split, i.e. leads to subsets that tend to contain examples of a single class.
Here we will assume binary features.

Input:  $\mathcal{D}$ is a labeled set of examples $(\mathbf{x}_i, y_i)$

*generate_tree*($\mathcal{D}$):

* if stopping criterion:  return
* pick best feature according to the impurity criterion
* define $\mathcal{D}_0$ as the set of examples for which best feature = 0
* define $\mathcal{D}_1$ as the set of examples for which best feature = 1
* recursively generate two subtrees using $\mathcal{D}_0$ and $\mathcal{D}_1$

To fully specify the algorithm we need to decide on a stopping criterion, and a way of choosing good features for performing a split on.  Split criteria are based on the idea of looking for splits that decrease the "impurity" in a node in the decision tree, i.e. make it so that the examples are more homogeneous.  Examples of measures that are used to quantify impurity include the Gini impurity and  information gain.

When growing a decision tree you can stop when one of the following criteria is satisfied:

* if maximum depth has been reached
* if leaf node contain only examples of the same class
* if leaf node contains very few examples

Decision trees can be extended with the option for pruning the tree after it has been created to remove nodes deep in the tree that can lead to overfitting.  They can also be extended to perform regression (cf. the CART decision tree algorithm).

### Gini impurity

The decision tree algorithm relies on a measure of impurity.  The Gini measure of impurity is a commonly used option.  It's defined as follows.
Let's assume a dataset with $C$ classes where at a given node they occur with probabilities $p_1,\ldots,p_C$.  The Gini impurity can now defined as:
$$
\text{Gini}(p) = 1 - \sum_{i=1}^C  p_i^2.
$$

Analyze the behavior of the Gini coefficient when you have two classes to convince yourself that it is indeed a measure of impurity.


Heres how Gini impurity is used to select a split:

* **Calculate Gini impurity for each split**: For each possible split on a feature, calculate the Gini impurity of the two resulting child nodes.

* **Compute the Weighted Average**: For each split, the Gini impurity of the left and right child nodes is weighted by the proportion of samples that fall into each child node.  The formula for weighted Gini impurity is:
$$
   \text{Gini(split)} = \frac{N_{\text{left}}}{N} \cdot \text{Gini}_{\text{left}} + \frac{N_{\text{right}}}{N} \cdot \text{Gini}_{\text{right}}
$$
where $N$ is the total number of examples in the node being split, and $N_{\text{left}}$ and $N_{\text{right}}$ are the number of examples in the left and right child nodes.

* **Choose the split with the lowest Gini impurity**.


### Advantages and disadvantages of decision trees

#### Advantages

* Simple to understand and interpret, even for a non-expert.
* Whereas most machine learning techniques are "black boxes", decision trees are more of a "white box".
* Easily handle a mix of categorical and numerical data.
* Can handle missing data.
* Fast training.

#### Disadvantages

* Decision trees are unstable, meaning that a small change in the data can lead to a large change in the structure of the model.  This calls into question the interpretation of the model.
* They are often relatively inaccurate, especially for data that is high dimensional. There are other methods that perform better (e.g. SVMs).

Decision trees have been among the most popular classification algorithms because of their speed and versatility.  Their biggest drawback has been their relatively low accuracy.
Random forests are a subsequent development that provide a big improvement in accuracy while maintaining most of their advantages.
